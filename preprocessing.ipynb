```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download("stopwords")

stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|@\S+|#\S+|[^a-z\s]", "", text)
    tokens = text.split()   # ðŸ”¥ SIMPLE TOKENIZATION
    tokens = [stemmer.stem(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

# Load CSVs without headers
train_df = pd.read_csv("twitter_training.csv", header=None)
val_df = pd.read_csv("twitter_validation.csv", header=None)

columns = ["tweet_id", "entity", "sentiment", "text"]
train_df.columns = columns
val_df.columns = columns

train_df["clean_text"] = train_df["text"].apply(clean_text)
val_df["clean_text"] = val_df["text"].apply(clean_text)

train_df.to_csv("train_clean.csv", index=False)
val_df.to_csv("val_clean.csv", index=False)

print("Preprocessing completed successfully")
```
